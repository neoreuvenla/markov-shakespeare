```python
import glob
import random
import re


class Poem:

    # Init

    def __init__(self, order):
        self.order = order  # number of previous words to read
        self.group_size = self.order + 1  # previous and current word
        self.text = ""
        self.graph = {}
        self.read()
        self.clean()
        self.train()
        #return  # needed?
        self.check = ""

    # Read

    def read(self):

        files = glob.glob("books/*.txt")

        for item in files:
            with open(item, "r") as opened:
                self.text = self.text + " " + opened.read()

    # Clean

    def clean(self):

        # list of regex expressions to remove from imported text
        edits = [
            "\n",  # removes new lines
            "\[.*?\]",  # removes stage directions
            "\W{2,}",  # removes punctuation
            "(?<!\n)[A-Z][a-z]{4,}('[a-z]|)",  # attempts to remove proper nouns
            "(ACT|Scene)\s\d",  # removes the text act/scene and number
            "[A-Z][A-Z]+",  # removes capitalised words
            "\[.*?\]",  # removes stage directions
            "\W{2,}",  # removes inter-word punctuation
            "\d",  # removes any numbers not otherwise removed
            "(\b[a-zA-Z]+'d|\b[a-zA-Z]+'[a-zA-Z]+\b|'\b[a-zA-Z]+\b|\b[a-zA-Z]')",  # removes abbreviations
        ]

        for i in range(len(edits)):
            self.text = re.sub(edits[i], " ", self.text, flags=re.S).strip()

    # Train

    def train(self):
        self.check = self.text 
        self.text = self.text.lower().split()
        self.text = self.text + self.text[: self.order]

        for i in range(0, len(self.text) - self.group_size):
            key = tuple(self.text[i : i + self.order])
            value = self.text[i + self.order]

            if key in self.graph:
                self.graph[key].append(value)
            else:
                self.graph[key] = [value]

    # Generate

    def generate(self, length):

        index = random.randint(0, len(self.text) - self.order)
        result = self.text[index : index + self.order]
        line = ""

        for i in range(length):

            state = tuple(result[len(result) - self.order :])
            next_word = random.choice(self.graph[state])
            result.append(next_word)

        for i in result:
            line += " " + i

        return line

    # Write

    def write(self):

        lines = random.randint(1, 3) #<<<<<<<<<<<# number of lines

        poem = ""
        edits = ["(?<=\n)\W", "\A([a-z]{1,2}\W)+", "(\W[a-z]{1,4})+\Z",  ]  # removes white spaces after new lines

        for i in range(lines):
            draft = ""

            while re.search(draft, self.check) != None:
                draft = self.generate(random.randint(5, 10))
                #draft = self.generate(8) #<<<<<<<<<<<<<<
            

         
            
            
            
            
            
            for item in edits:
                draft = re.sub(item, "", draft).strip()
            
            draft = draft.capitalize() # captialises first letter of each line before adding
            poem += "\n" + draft

        for item in edits:
            poem = re.sub(item, "", poem).strip()
        
       
        print(poem)  # for printing text
        #return poem #use for printing to twitter
```
